# Data Transformation Assignment

```python
# Import Libraries
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Load Dataset

df = pd.read_csv("/content/employee_productivity_dataset.csv")
```

```python
# Display first 5 rows
print(df.head())
```

```python
# Q1. Handle Missing Values (Basic Cleaning)

# Fill missing values using the following:
# - Age -> Median
# - Salary -> Mean
# - Hours_Worked_Per_Week -> Median
# - Performance_Score -> Mean

# Display the dataset after handling missing values.

# Step 1: Check Missing Values
print(df.isnull().sum())
```

```python
# Step 2: Fill Missing Values

# Fill Age with Median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill Salary with Mean
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

# Fill Hours_Worked_Per_Week with Median
df['Hours_Worked_Per_Week'] = df['Hours_Worked_Per_Week'].fillna(
    df['Hours_Worked_Per_Week'].median()
)

# Fill Performance_Score with Mean
df['Performance_Score'] = df['Performance_Score'].fillna(
    df['Performance_Score'].mean()
)
```

```python
# Display Dataset After Cleaning

print(df.head())
```

## Explanation

Missing values can affect machine learning model performance.

We replace missing values using:
- Median is used for Age and Hours_Worked_Per_Week because it handles outliers better.
- Mean is used for Salary and Performance_Score because numerical averages are suitable here.



```python
# Q2. Label Encoding

# Covert these columns into numeric values using Label Encoding:
# - Gender
# - Department

# Show the updated columns.

# Step 1: Create Label Encoder

le = LabelEncoder()
```

```python
# Step 2: Apply Label Encoding

df['Gender'] = le.fit_transform(df['Gender'])

df['Department'] = le.fit_transform(df['Department'])
```

```python
# Step 3: Display Updated Columns

print(df[['Gender', 'Department']].head())
```

## Explanation

- Machine learning models cannot understand text values directly.
- Label Encoding converts categorical text data into numerical form.

```python
# Q3. One-Hot Encoding

# Apply One-Hot Encoding on:
# - Work_Mode
# - Location

# Display the dataset and check how many new columns are created.

# Step 1: Apply pd.get_dummies()

df = pd.get_dummies(df,
                    columns=['Work_Mode', 'Location'],
                    drop_first=True)
```

```python
# Step 2: Display Dataset

print(df.head())
```

```python
# Step 3: Check Number of New Columns

print("Total Columns After Encoding:", df.shape[1])
```

## Explanation

- One-Hot Encoding converts categories into separate binary columns.

```python
# Q4. Normalization (Min-Max Scaling)

# Normalize the following columns:
# - Salary
# - Hours_Worked_Per_Week

# Use MinMaxScaler and display results.

# Step 1: Create Scaler

minmax = MinMaxScaler()
```

```python
# Step 2: Apply Scaling

df[['Salary', 'Hours_Worked_Per_Week']] = minmax.fit_transform(
    df[['Salary', 'Hours_Worked_Per_Week']]
)
```

```python
# Step 3: Display Results

print(df[['Salary', 'Hours_Worked_Per_Week']].head())
```

## Explanation

- Normalization scales values between 0 and 1.

- It is useful when features have different ranges.


```python
# Q5. Standardization (Scaling)

# Apply StandardScaler on:
# - Age
# - Projects_Completed

# Display the transformed values

# Step 1: Create StandardScaler

standard = StandardScaler()
```

```python
# Step 2: Apply Standardization

df[['Age', 'Projects_Completed']] = standard.fit_transform(
    df[['Age', 'Projects_Completed']]
)
```

```python
# Step 3: Display Transformed Values

print(df[['Age', 'Projects_Completed']].head())
```

## Explanation

Standardization changes data so that:
- Mean = 0
- Standard Deviation = 1

It is useful for algorithms like:
- Logistic Regression
- KNN
- SVM

```python
# Q6. Compare Scaling Methods

# Apply both:
# - MinMaxScaler
# - StandardScaler

# On the Salary column.
# Show both results side by side.

# Step 1: Create New Columns

# MinMax Scaling
df['Salary_MinMax'] = MinMaxScaler().fit_transform(df[['Salary']])

# Standard Scaling
df['Salary_Standard'] = StandardScaler().fit_transform(df[['Salary']])
```

```python
# Step 2: Display Both Results

print(df[['Salary', 'Salary_MinMax', 'Salary_Standard']].head())
```

## Explanation

Both scaling methods transform numerical data differently.

MinMaxScaler
- Converts values between 0 and 1
- Sensitive to outliers

StandardScaler
- Converts data around mean 0
- Handles outliers better

```python
# Q7. Build Preprocessing Pipeline

# Create a pipeline that:
# - Applies encoding to categorical columns
# - Applies scaling to numerical columns

# Use:
# - ColumnTransformer
# - Pipeline

# Step 1: Reload Original Dataset

df = pd.read_csv("/content/employee_productivity_dataset.csv")
```

```python
# Step 2: Define Columns

categorical_cols = ['Gender',
                    'Department',
                    'Work_Mode',
                    'Location']

numerical_cols = ['Age',
                  'Salary',
                  'Hours_Worked_Per_Week',
                  'Performance_Score',
                  'Projects_Completed']
```

```python
# Step 3: Create Transformers

preprocessor = ColumnTransformer(
    transformers=[

        ('num', StandardScaler(), numerical_cols),

        ('cat', OneHotEncoder(drop='first'), categorical_cols)

    ]
)
```

```python
# Step 4: Create Pipeline

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
])
```

## Explanation

A pipeline automates preprocessing steps.

Benefits:
- Cleaner code
- Faster workflow
- Reduces errors
- Easy reuse in ML projects

ColumnTransformer helps apply:
- Scaling on numerical columns
- Encoding on categorical columns

```python
# Q8. Apply Pipeline

# Apply the pipeline on the dataset.

# Display:
# - Trasnformed dataset
# - Shape of final dataset

# Step 1: Transform Dataset

transformed_data = pipeline.fit_transform(df)
```

```python
# Step 2: Display Transformed Dataset

print(transformed_data)
```

```python
# Step 3: Display Shape

print("Shape of Final Dataset:", transformed_data.shape)
```

## Explanation

The pipeline performs all preprocessing steps automatically in one command.

It:
- Encodes categorical columns
- Scales numerical columns
- Produces final transformed dataset
